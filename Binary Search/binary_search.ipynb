```python
import math
import matplotlib.pyplot as plt

class BinarySearch:
    def __init__(self, nums):
        self.nums = sorted(nums)

    def __call__(self, target):
        """
        Search method which try to find given target number.
        """
        count = 0
        left = 0
        right = len(self.nums) - 1

        while left <= right:
            mid = (left + right) // 2
            count += 1

            if self.nums[mid] == target:
                return {"index": mid, "steps": count, "found": True}
            elif self.nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1

        return {"index": -1, "steps": count, "found": False}

    @staticmethod
    def plot_time_complexity(max_n=1024):
        """
        Visualizes how the number of steps grows logarithmically.
        """
        n_values = []
        steps = []

        n = 2
        while n <= max_n:
            n_values.append(n)
            steps.append(math.log2(n))
            n *= 2

        plt.figure(figsize=(10, 5))
        plt.plot(n_values, steps, marker='o', linestyle='-', color='green')
        plt.xlabel("Input size (n)")
        plt.ylabel("Maximum Steps ($log_2 n$)")
        plt.title("Binary Search Time Complexity: $O(\log n)$")
        plt.grid(True)
        plt.show()


data = list(range(1, 101))
searcher = BinarySearch(data)


print(f"Search Result: {searcher(42)}")


BinarySearch.plot_time_complexity(2048)
```
